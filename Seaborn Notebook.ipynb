```python
#Pair Plot

sns.pairplot(df,hue= 'Global_Sales')
plt.show()
```

```python
# Discretize the sales columns into bins
df["EU_Sales_Binned"] = pd.cut(df["EU_Sales"], bins=5, labels=["Very Low", "Low", "Medium", "High", "Very High"])
df["JP_Sales_Binned"] = pd.cut(df["JP_Sales"], bins=5, labels=["Very Low", "Low", "Medium", "High", "Very High"])

# FacetGrid with binned columns
g = sns.FacetGrid(df, col="EU_Sales_Binned", row="JP_Sales_Binned")
g.map(sns.scatterplot, "Global_Sales", "Other_Sales")
plt.show()
```

```python
#Full Workflow of the Video Game Sales
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
#Reading the CSV file
data = pd.read_csv('vgsales.csv')
```

```python
# Visualizations
plt.figure(figsize=(12, 6))
sns.scatterplot(x="Global_Sales", y="Other_Sales", hue="Genre", data=data)
plt.title("Video Game Sales Data")
plt.xlabel("Global Sales")
plt.ylabel("Other Sales")
plt.show()
```

```python
#Since the Data are in object and Category type 
data = data.drop(columns=["Name", "Platform","Genre",'Publisher'])
```

```python
#Correlation HeatMap for Video Game Sales Data
corr = data.corr()
sns.heatmap(corr,annot=True,cmap='coolwarm',fmt='.2f') #fmt = format , annot = Values are displayed inside the Cells
plt.title('Correlation HeatMap for the Video Game Sales Data ')
plt.show()
```

```python
#Visualizing  pairwise relationships between numerical variables

sns.pairplot(df[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Global_Sales']], diag_kind='kde')
plt.suptitle("Pairplot of Sales Data", y=1.02)
plt.show()
```

```python
#Regression Analysis 
sns.regplot(data=df, x='NA_Sales', y='Global_Sales')
plt.title("Regression Plot: NA Sales vs Global Sales")
plt.show()
```

```python
#Regression Fit
sns.residplot(data=df, x='NA_Sales', y='Global_Sales')
plt.title("Residual Plot")
plt.show()
```
